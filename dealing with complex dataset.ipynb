## This dataset has 1037 rows with 13 columns,, and all columns has dtype object, expect for Founded Year, Investors Count, Deal Terms, and Portfolio Exits  has type floats
- first step we need to convert columns (numbers in Valuation (```$```B), and Total Raised) to the real number, EX:insted of  ```$```140 B i want to have 140000000000
- Second step removing the ```$``` sign from the title, Valuation (```$```B) will become Valuation (```$```)

````python
#having a copy to make to compare after transforming
df_copy = df.copy()
````

````python
#cahning the name of our columns using the .rename() method
df_copy.rename(columns = {'Valuation ($B)':'Valuation ($)'},inplace=True)
df_copy.rename(columns = {'Total Raised':'Total Raised ($)'},inplace=True)
#having a list of all our columns names will be usfull 
col_names = df_copy.columns.tolist()
col_names
````

````python
#checking 
df_copy.head()
````

````python
#removing the dollar sign from the  Valuation ($) , Total Raised ($) columns
df_copy['Valuation ($)'] = df_copy['Valuation ($)'].apply(lambda x: x.replace('$',''))
df_copy['Total Raised ($)'] = df_copy['Total Raised ($)'].astype(str)
df_copy['Total Raised ($)'] = df_copy['Total Raised ($)'].apply(lambda x: x.replace('$',''))
#we did not remove the $ from Total Raised ($) because it was taking care of when it was converted to floats 
````

````python
df_copy.head()
````

````python
#function that takes the str number , decide if it's million or buillion, from the letter,and add the coresponding zeros to it 
def change_m_b(input_num): 
    num = str(input_num)
    for i in range(len(num)): 
        if num[i] == 'B':
            float_num = float(num[:i])
            float_num = float_num * 1000000000
            return int(float_num)
        elif num[i] == 'M':
            float_num = float(num[:i])
            float_num = float_num * 1000000
            return int(float_num)
        else :
            pass 
````

````python
#testing if the conversion code wroks as expected or not 
print(f"original value : {df_copy['Total Raised ($)'][1032]}, after chaning : {change_m_b(df_copy['Total Raised ($)'][1032])}")
````

````python
#applying the code to the Total Raised columns
for i in range(df.shape[0]):
    df_copy['Total Raised ($)'][i] = change_m_b(df_copy['Total Raised ($)'][i])  
````

````python
# now doing the same for the Valuation
def change_b(input_num):
    num = str(input_num)
    for i in range(len(num)): 
            float_num = float(num)
            float_num = float_num * 1000000000
            return int(float_num)
````

````python
#applying the code to the Valuation column
for i in range(df.shape[0]):
    df_copy['Valuation ($)'][i] = change_b(df_copy['Valuation ($)'][i])  
````
